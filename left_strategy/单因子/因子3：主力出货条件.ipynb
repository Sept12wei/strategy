## 主力出货条件统计分析

#### 获取1000支股票数据

```python
import numpy as np
import pandas as pd
import yfinance as yf
```

```python
df = pd.read_csv('./data/all_data_SS_2y.csv')
df = df.drop(df.columns[0], axis=1)
df
```

```python
# 过滤掉code列中以'300'、'688'、‘8’开头的股票
df = df[~df['code'].str.startswith('300')]
df = df[~df['code'].str.startswith('688')]
df = df[~df['code'].str.startswith('8')]
df
```

```python
df.set_index('code', inplace=True)
df
```

```python
df['Pct_Chg'] = df.groupby(level=0)['Close'].pct_change()
df['Pct_Chg_5d'] = df.groupby(level=0)['Close'].pct_change(periods=5)
df = df.dropna(subset=['Pct_Chg_5d'])
df
```

```python
stock_data = df.index.get_level_values('code').unique()
stock_data = pd.DataFrame({
    'code': stock_data,
    'is_crash': 0,  # 初始化is_crash列为0
    'is_low': 0
})
stock_data
```

### 这里设置前期暴跌的条件是5日跌20%

```python
df = df.reset_index()
# 遍历df中的每一行
for index, row in df.iterrows():
    # 检查Pct_Chg_5d是否小于-0.2
    if row['Pct_Chg_5d'] < -0.2:
        # 如果是，更新stock_data中的is_crash为1
        stock_data.loc[stock_data['code'] == row['code'], 'is_crash'] = 1
stock_data
```

```python
count_1 = (stock_data['is_crash'] == 1).sum()
count_0 = (stock_data['is_crash'] == 0).sum()
print(f"Number of 1's: {count_1}")
print(f"Number of 0's: {count_0}")
```

### 有398支股票经历过大跌

## 长期慢跌的判断

#### 我认为，如果在过去90天的交易日内，存在60天及以上的数据符合 MA5 < MA10 < MA20 < MA30 ，则可以认为这只股票处于长期的下跌趋势中

```python
df['MA5'] = df['Close'].rolling(window=5).mean()
df['MA10'] = df['Close'].rolling(window=10).mean()
df['MA20'] = df['Close'].rolling(window=20).mean()
df['MA30'] = df['Close'].rolling(window=30).mean()
df = df.dropna(subset=['MA5','MA10','MA20','MA30'])
df
```

```python
df['Condition'] = (df['MA5'] < df['MA10']) & (df['MA10'] < df['MA20']) & (df['MA20'] < df['MA30'])
df['Condition'] = df['Condition'].astype(int)
df
```

```python
df['low'] = df['Condition'].rolling(window=90).mean()
df_low = df[df['low'] > 0.6]
df_low
```

```python
df_low = df_low.reset_index()
# 遍历df中的每一行
for index, row in df_low.iterrows():
    stock_data.loc[stock_data['code'] == row['code'], 'is_low'] = 1
stock_data
```

```python
count_1_low = (stock_data['is_low'] == 1).sum()
count_0_low = (stock_data['is_low'] == 0).sum()
print(f"Number of 1's: {count_1_low}")
print(f"Number of 0's: {count_0_low}")
```

```python
stock_data_condition = stock_data[(stock_data['is_crash'] == 1) & (stock_data['is_low'] == 1)]
stock_data_condition
```

#### 一共有108支股票符合前期暴跌并且长期慢跌的条件
